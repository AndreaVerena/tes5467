## Introducción
El presente proyecto corresponde a una aplicación que está pensada con la visión de procesar imágenes aéreas capturadas por drones, a fin de identificar incendios forestales. Para fines de este curso, se propone:
* Entrenar un modelo de red neuronal que permita clasificar imágenes aéreas en dos categorías: "Fire" y "No_Fire", que corresponden a si se detecta o no un incendio en las imágenes.
* Disponibilizar el modelo a través de una aplicación web, donde el usuario deberá cargar manualmente una imagen y se entregará como respuesta la categoría a la que pertenece esa imagen según la predicción generada por el modelo.

## Desarrollo
### Recolección de imágenes

````python
from fastai.vision import *
````

Para conseguir las imágenes se utiliza la metodología vista en el curso. Se visita [Google Images](http://images.google.com), se buscan imágenes con "aerial shots forest -fire" para obtener imágenes aéreas de bosques para la clase "nofire", y por otra parte, imágenes con "" para las imágenes de la categoría "fire".

Luego de hacer scroll sobre la búsqueda en el navegador Google Chrome, se presiona la combinación <kbd>Ctrl</kbd><kbd>Shift</kbd><kbd>j</kbd> y se ejecuta el siguiente comando para obtener un archivo con las URLs de las imágenes que fueron scrolleadas:  

Se crean los dirctorios con los nombres de las categorías:

````python
folder = 'Fire'
file = 'Fire.csv'
````

````python
folder = 'No_Fire'
file = 'No_Fire.csv'
````

````python
path = Path('data')
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
````

````python
path.ls()
````

````python
classes = ['Fire','No_Fire']
````

Se descargan las imágenes que fueron scrolleadas y guardadas en el archivo "Fire.csv", primero para la clase "Fire" y luego se repite para la clase "No_Fire". Se utiliza el parámetro **max_workers = 0** porque de lo contrerio se muestra error. Luego se realiza una selección manual para escoger 200 imágenes eliminando aquellas que no correspondan a la descripción de la clase.

````python
#download_images(path/file, dest, max_workers=0)
````

````python
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
````

````python
np.random.seed(201)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=0).normalize(imagenet_stats)
````

````python
data.classes
````

````python
data.show_batch(rows=3, figsize=(7,8))
````

A continuación se muestran las clases, la cantidad de clases, el tamaño del conjunto de entrenamiento (320 imágenes) y el tamaño del conjunto de validación (80 imágenes), es decir, 80% entrenamiento y 20% validación.

````python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
````

### Entrenamiento del modelo
Para entrenar el modelo se utiliza la red neuronal convolucional resnet34 vista en el curso. El modelo se entrena con 4 épocas.

````python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
learn.model
````

````python
learn.fit_one_cycle(4)
````

Se observa que en las últimas dos iteraciones el modelo logra disminuir la tasa de error a un 2,5% y finalmente a 0,0%.

````python
learn.save('stage-1')
````

### Resultados

````python
interp = ClassificationInterpretation.from_learner(learn)
losses,idxs = interp.top_losses()
len(data.valid_ds)==len(losses)==len(idxs)
````

````python
interp.plot_top_losses(9, figsize=(15,11))
````

#### Matriz de confusión

````python
interp.plot_confusion_matrix()
````

De acuerdo a lo que muestra la matriz, de las 80 imágenes que fueron entregadas al modelo para la validación, **80 fueron predichas correctamente** y el modelo no **erró en 2 imágenes**, que eran de la categoría No_Fire, pero que el modelo los predijo como "Fire". En este caso se trataría de 2 "falsos positivos" ya que la clase de interés son los "Fire" (clase positiva).
* Accuracy:    97.5%
* Recall:     100.0% 
* Precision:   95.1%
